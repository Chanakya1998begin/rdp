# Opening-Audit LLM Advisory — NVIDIA DGX Cloud (Colab runner)

Re-runs the **opening-audit advisory** (the `STRONG_BEARISH / BULLISH-LEAN / MIXED` verdict the trapX engine emits right after the 09:19 bar closes) on a chosen day's data, using NVIDIA's **DGX Cloud chat-completions API** (`integrate.api.nvidia.com/v1`) via the OpenAI Python SDK.

**How it works**

| # | Stage | Description |
|---|---|---|
| 1 | Setup | Read `NVIDIA_API_KEY` from Colab Secrets; OpenAI SDK already shipped with Colab |
| 2 | Skill | Clone trapX (private) using `GITHUB_TOKEN` Colab Secret; load `opening_audit_summary_v2.md` |
| 3 | Drive | Mount Google Drive; locate the day's `trapx_*.log` |
| 4 | Helpers | Inline parsers + NVIDIA call + 3-line verdict parser + rendered-block formatter |
| 5 | Inspect | Pull the `🔍 OPEN AUDIT · <date> · <time>` block from the log |
| 6 | Run | 9-stage trace: NVIDIA preflight → build user message → POST → parse → render → save |
| 7 | Save | Dump `audit_record` (JSON) back to Drive — accumulates alongside production records |

**Prerequisites — Colab Secrets**

| Key | Scope | Purpose |
|---|---|---|
| `NVIDIA_API_KEY` | — | DGX Cloud gateway auth. Format `nvapi-...`. |
| `GITHUB_TOKEN` | `repo` | Private-repo clone of `Chanakya1998begin/trapX`. |

**Why NVIDIA backend instead of Ollama?**

- No 2-GB model download → faster cold-start
- Llama 3.3 70B Instruct ≫ Llama 3.2 3B for nuanced V2 reasoning
- Cloud serverless, no GPU runtime required
- Matches the third LLM-advisory backend shipped in trapX CHA-173

## 4. Helpers — parsing, NVIDIA call, render

All inlined so the notebook is self-contained. The 3-line parser + renderer mirror `project/llm_advisory/advisory.py` so the rendered output is byte-identical to production rendering.

````python
import json
import re
import textwrap
import time
import datetime as _dt
from openai import OpenAI

# ---- IST timestamp ---------------------------------------------------

def now_ist() -> str:
    return (_dt.datetime.utcnow() + _dt.timedelta(hours=5, minutes=30)).strftime("%H:%M:%S")

# ---- Log-file parsing -----------------------------------------------
#
# Two log formats to handle:
#
#   LIVE mode (no prefix):
#     🔍 OPEN AUDIT · 18-May · 09:20
#     ━━━━━━━━━━━━━━━━━━━━━━
#     🏁 STRONG_BEARISH  🔴 (-2)
#     ...
#     ━━━━━━━━━━━━━━━━━━━━━━
#
#   FAST-FORWARD replay mode (each line prefixed `│ `):
#     ┌─ suppressed body ─
#     │ 🔍 OPEN AUDIT · 18-May · 09:20
#     │ ━━━━━━━━━━━━━━━━━━━━━━
#     │ ...
#     └─

_OA_HEADER_RE = re.compile(
    r"^[│\s]*🔍 OPEN AUDIT · (?P<date>\d+-[A-Z][a-z]+) · (?P<time>\d+:\d+)\s*$"
)
_SEP_RE = re.compile(r"^[│\s]*━{20,}\s*$")
_LINE_PREFIX_RE = re.compile(r"^\s*│ ?")


def extract_opening_audit_body(log_path: str) -> dict:
    """Find the first `🔍 OPEN AUDIT · <date> · <time>` block followed
    by a `━━━` separator, then capture all lines through the closing
    `━━━`. Strips the `│ ` prefix used in fast-forward replay so the
    returned `body` is uniform.

    Returns {date, time, body, lines}.
    """
    with open(log_path, encoding="utf-8") as f:
        lines = f.readlines()
    for i, ln in enumerate(lines):
        m = _OA_HEADER_RE.match(ln.rstrip("\n"))
        if not m:
            continue
        if i + 1 >= len(lines) or not _SEP_RE.match(lines[i + 1].rstrip("\n")):
            continue
        captured = [lines[i].rstrip("\n"), lines[i + 1].rstrip("\n")]
        j = i + 2
        while j < len(lines):
            captured.append(lines[j].rstrip("\n"))
            if _SEP_RE.match(lines[j].rstrip("\n")):
                break
            j += 1
        stripped = [_LINE_PREFIX_RE.sub("", ln_) for ln_ in captured]
        body = "\n".join(stripped).strip()
        return {
            "date":  m.group("date"),
            "time":  m.group("time"),
            "body":  body,
            "lines": len(stripped),
        }
    raise RuntimeError(
        f"No `🔍 OPEN AUDIT · ... \n━━━ ... ━━━` block in {log_path}. "
        "The opening-audit may not have fired (engine startup before 09:20?)"
    )


def build_user_message_from_body(body_info: dict) -> str:
    """Build the v2 prompt with the extracted formatted body as input."""
    return (
        "Apply the structural-framework rules to this opening-audit "
        "snapshot and output the 3-section advisory (Verdict / Action / Dtls) "
        "per the v2 contract. The snapshot below is the formatted OPEN AUDIT "
        f"body the trapX engine printed at {body_info['date']} "
        f"{body_info['time']}:\n\n"
        "```\n"
        f"{body_info['body']}\n"
        "```"
    )


# ---- NVIDIA chat-completions call ------------------------------------

def call_nvidia(skill: str, user_msg: str, *, model: str = MODEL,
                api_key: str = NVIDIA_API_KEY,
                base_url: str = NVIDIA_BASE_URL,
                temperature: float = 0.0,
                timeout_sec: int = 120) -> dict:
    """Mirrors project/llm_advisory/client.py:_call_nvidia. POSTs to the
    DGX Cloud chat-completions endpoint with the (system, user) message
    pair. Returns the normalized {text, usage, latency_ms, stop_reason}
    shape used elsewhere in the trapX advisory layer.

    NO `max_tokens` cap (matches CHA-173 follow-up): replay / A-B-test
    usage wants the complete skill output, not artificial truncation.
    """
    client = OpenAI(
        base_url=base_url,
        api_key=api_key,
        timeout=timeout_sec,
    )
    t0 = time.perf_counter()
    # CHA-185 backport of CHA-174: seed=42 locks the NVIDIA gateway to a
    # deterministic response on identical input. Without it, NVIDIA can
    # produce different scores (and even flip the sign) on the same
    # snapshot + skill — observed today across 3 consecutive runs.
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": skill},
            {"role": "user",   "content": user_msg},
        ],
        temperature=temperature,
        seed=42,
    )
    latency_ms = round((time.perf_counter() - t0) * 1000, 1)
    text = (response.choices[0].message.content or "").strip()
    usage = response.usage
    return {
        "text": text,
        "usage": {
            "input_tokens":  getattr(usage, "prompt_tokens", 0) or 0,
            "output_tokens": getattr(usage, "completion_tokens", 0) or 0,
        },
        "latency_ms": latency_ms,
        "stop_reason": getattr(response.choices[0], "finish_reason", "") or "",
        "model":       getattr(response, "model", model),
        "id":          getattr(response, "id", ""),
    }


# ---- 3-line parser (mirrors advisory.py) -----------------------------

_BIAS_EMOJIS = ("🟢", "🔴", "🟡", "⚪")
# CHA-PATCH: verdict-style touchpoints (and opening-audit on model drift)
# emit ✅/⚠️/❌/🔄 instead of bias emojis. Accept these as a
# fallback in extract_bias_line so a non-bias leading emoji doesn't
# silently produce empty render → empty log file.
_VERDICT_EMOJIS = ("✅", "⚠️", "❌", "🔄")
_ALL_LEAD_EMOJIS = _BIAS_EMOJIS + _VERDICT_EMOJIS
_SCORE_PAT = re.compile(
    r"(?:📊\s*)?Score\s*:[^+\-\d]*([+-]?\d+(?:\.\d+)?)",
    re.IGNORECASE,
)


def extract_bias_line(raw: str) -> str:
    """CHA-patched: 3-pass extractor.

    Pass 1 — line starts with a bias emoji (opening-audit canonical).
    Pass 2 — line starts with a verdict emoji (model drift fallback).
    Pass 3 — any known lead emoji anywhere in the raw text → return the
             first non-fence line (most permissive fallback).
    """
    if not raw:
        return ""
    _FENCE = "```"
    # Pass 1: bias-emoji prefix (canonical)
    for line in raw.splitlines():
        s = line.strip().lstrip(">").strip()
        if s.startswith(_FENCE):
            continue
        if any(s.startswith(e) for e in _BIAS_EMOJIS):
            return s
    # Pass 2: verdict-emoji prefix (CHA model-drift fallback)
    for line in raw.splitlines():
        s = line.strip().lstrip(">").strip()
        if s.startswith(_FENCE):
            continue
        if any(s.startswith(e) for e in _VERDICT_EMOJIS):
            return s
    # Pass 3: any known emoji anywhere → first non-fence line
    if any(e in raw for e in _ALL_LEAD_EMOJIS):
        for line in raw.strip().splitlines():
            if not line.startswith(_FENCE):
                return line.strip()
    return ""



def extract_score(raw: str):
    for line in (raw or "").splitlines():
        stripped = line.strip().lstrip(">").strip()
        if stripped.startswith("```"):
            continue
        m = _SCORE_PAT.search(stripped)
        if not m:
            continue
        try:
            v = float(m.group(1))
        except (TypeError, ValueError):
            continue
        return max(-1.0, min(1.0, v))
    return None


def extract_action(raw: str) -> str:
    if not raw:
        return ""
    inline_pat = re.compile(r"(?:🎯\s*)?Action\s*:\s*(.+)", re.IGNORECASE)
    label_only_pat = re.compile(r"(?:🎯\s*)?Action\s*:\s*$", re.IGNORECASE)
    section_break = re.compile(
        r"^(?:📊\s*)?(?:Verdict|Score|Dtls)\s*:", re.IGNORECASE,
    )
    lines = raw.splitlines()
    for i, line in enumerate(lines):
        stripped = line.strip().lstrip(">").strip()
        if stripped.startswith("```"):
            continue
        m = inline_pat.match(stripped)
        if m:
            text = m.group(1).strip()
            if len(text) >= 2 and text[0] in '"\'' and text[-1] == text[0]:
                text = text[1:-1].strip()
            if text:
                return text
        if label_only_pat.match(stripped):
            bullets = []
            for nxt in lines[i + 1:]:
                ns = nxt.strip().lstrip(">").strip()
                if not ns:
                    break
                if ns.startswith("```") or section_break.match(ns):
                    break
                if ns and ns[0] in "•-*":
                    ns = ns[1:].strip()
                if ns:
                    bullets.append(ns)
            if bullets:
                joined = [b if b.endswith((".", "!", "?")) else b + "." for b in bullets]
                return " ".join(joined)
    return ""


# ---- render_block (mirrors _render_advisory_block) -------------------

_LLM_SEPARATOR = "━" * 22
_LINE_MAX_WIDTH = 40


def _wrap_long_line(text: str, width: int = _LINE_MAX_WIDTH):
    if not text or not text.strip():
        return [text]
    return textwrap.wrap(text, width=width,
                         break_long_words=False, break_on_hyphens=False) or [text]


def _split_action_sentences(action_text: str):
    if not action_text or not action_text.strip():
        return []
    parts = re.split(r"(?<=[.!?])\s+", action_text.strip())
    return [p.strip() for p in parts if p.strip()]


def _strip_bias_emoji(line: str) -> str:
    for e in _BIAS_EMOJIS:
        if line.startswith(e):
            return line[len(e):].lstrip()
    return line


def _bias_emoji_of(line: str) -> str:
    for e in _BIAS_EMOJIS:
        if line.startswith(e):
            return e
    return ""


def _format_score(value):
    if value is None:
        return ""
    if value == 0.0:
        return "0.00"
    return f"+{value:.2f}" if value > 0 else f"{value:.2f}"


def render_block(bias_line: str, score, action_text: str,
                 bar_time: str, day_stamp: str) -> str:
    if not bias_line:
        return ""
    emoji = _bias_emoji_of(bias_line)
    if not emoji and score is not None:
        emoji = ("🔴" if score <= -0.30 else
                 "🟡" if score <= -0.10 else
                 "⚪"  if score <  0.10 else
                 "🟡" if score <  0.30 else
                 "🟢")
    dtls = _strip_bias_emoji(bias_line)
    lines = [_LLM_SEPARATOR, "🤖 LLM advisory:"]
    score_str = _format_score(score)
    if score_str:
        bracketed = f"{emoji}[{score_str}]" if emoji else f"[{score_str}]"
        lines.append(f"Verdict: {bracketed}")
    if action_text:
        sentences = _split_action_sentences(action_text)
        if len(sentences) <= 1:
            lines.extend(_wrap_long_line(f"Action: {action_text.strip()}"))
        else:
            lines.append("Action:")
            for s in sentences:
                lines.extend(_wrap_long_line(f"• {s}"))
    lines.extend(_wrap_long_line(f"Dtls: {dtls}"))
    lines.append(f" --------- @ [{day_stamp} {bar_time}]")
    return "\n".join(lines)


print("✅ Helpers loaded:")
for h in [
    "extract_opening_audit_body(log_path)",
    "build_user_message_from_body(body_info)",
    "call_nvidia(skill, user_msg, model=...)",
    "extract_bias_line / extract_score / extract_action",
    "render_block(bias, score, action, bar_time, day_stamp)",
]:
    print(f"   • {h}")
````

## 6. Run the advisory on NVIDIA DGX Cloud

9-stage tracer. Each stage prints an IST timestamp + what just happened, so you can see where time goes.

````python
# ===== RUN CONFIG =====
TEMPERATURE  = 0.0   # Determinism
TIMEOUT_SEC  = 120   # Cloud cold-start can hit ~10-15s on the first call
# ======================
````

````python
def trace(stage, msg):
    print(f"[{now_ist()} IST] STAGE {stage}: {msg}")

print("━" * 70)
print(f"  OPENING-AUDIT LLM ADVISORY (NVIDIA)  ·  {MON_DD}  ·  bar_time={body_info['time']}")
print("━" * 70)

# STAGE 1 — preflight (we don't have a `tags` endpoint on NVIDIA gateway;
# the auth + model availability is verified by the actual call. Quick check:
# confirm key + base_url are populated.)
trace(1, f"NVIDIA backend: base_url={NVIDIA_BASE_URL}, model={MODEL}")
assert NVIDIA_API_KEY, "NVIDIA_API_KEY missing"
trace(1, "✅ key + base_url present (gateway auth checked on the actual call)")

# STAGE 2 — body already extracted in cell 5
trace(2, f"OPEN AUDIT body extracted from {pathlib.Path(LOG_FILE).name}")
trace(2, f"✅ {body_info['lines']} lines, date={body_info['date']}, time={body_info['time']}")

# STAGE 3 — build user message
trace(3, "Building user_message (skill = v2 opening_audit_summary_v2.md) ...")
user_msg = build_user_message_from_body(body_info)
trace(3, f"✅ user_message: {len(user_msg):,} chars")

# STAGE 4 — send to NVIDIA
trace(4, f"POST /v1/chat/completions — model={MODEL}, temp={TEMPERATURE}, "
         f"max_tokens=<unset, model default>")
t0 = time.perf_counter()
result = call_nvidia(SKILL, user_msg, model=MODEL,
                     temperature=TEMPERATURE, timeout_sec=TIMEOUT_SEC)
elapsed = time.perf_counter() - t0
trace(4, f"✅ Response in {elapsed:.1f}s "
         f"(in/out tokens={result['usage']['input_tokens']:,}/{result['usage']['output_tokens']:,}, "
         f"stop={result['stop_reason']})")

# STAGE 5 — show raw
trace(5, f"Raw text ({len(result['text']):,} chars):")
print()
print("─── RAW LLM OUTPUT ───")
print(result['text'])
print("─── END RAW ───")
print()

# STAGE 6 — parse
trace(6, "Parsing bias / score / action ...")
bias_line   = extract_bias_line(result["text"])
score_val   = extract_score(result["text"])
action_text = extract_action(result["text"])
trace(6, f"  bias_line  = {bias_line!r}")
trace(6, f"  score      = {score_val}")
_act_preview = (action_text[:80] + '...') if len(action_text) > 80 else action_text
trace(6, f"  action     = {_act_preview!r}")

# STAGE 7 — render
trace(7, "Rendering polished advisory block ...")
rendered = render_block(bias_line, score_val, action_text, body_info['time'], DAY_STAMP)
print()
print("═" * 70)
print(f"  NVIDIA OPENING-AUDIT ADVISORY  ·  {MON_DD}  ·  {body_info['time']}")
print("═" * 70)
print(rendered if rendered else "(empty — parser found no bias line)")
print("═" * 70)
print()

# STAGE 8 — assemble audit_record (for STAGE 9 save)
trace(8, "Assembling audit_record ...")
audit_record = {
    "ts": _dt.datetime.utcnow().isoformat() + "Z",
    "notebook": "Opening_Audit_Advisory_NVIDIA",
    "mon_dd": MON_DD,
    "source_log": pathlib.Path(LOG_FILE).name,
    "bar_date": body_info['date'],
    "bar_time": body_info['time'],
    "skill_version": "v2",
    "skill_chars": len(SKILL),
    "backend": "nvidia",
    "model": result['model'],
    "nvidia_base_url": NVIDIA_BASE_URL,
    "request": {
        "user_message_chars": len(user_msg),
        "opening_audit_body": body_info['body'],
    },
    "response": {
        "raw_text": result['text'],
        "parsed": {"verdict": bias_line, "score": score_val, "action": action_text},
        "rendered_block": rendered,
        "stop_reason": result['stop_reason'],
        "completion_id": result['id'],
    },
    "usage": result['usage'],
    "latency_ms": int(elapsed * 1000),
}

# STAGE 9 — summary
trace(9, "Done. `audit_record` is in the notebook namespace.")
print()
print(f"  Latency : {elapsed:.2f}s")
print(f"  Tokens  : input={result['usage']['input_tokens']:,}  output={result['usage']['output_tokens']:,}")
print(f"  Stop    : {result['stop_reason']}")
print(f"  Model   : {result['model']}")
````
